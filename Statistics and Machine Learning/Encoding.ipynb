```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('/content/spotify.csv')
df.head(5)
```

```python
# Label Encoding (Encoding playlist_genre)

le = LabelEncoder()
df['playlist_genre'] = le.fit_transform(df['playlist_genre'])
df.head(5)
```

```python
mapping_df = pd.DataFrame({
    "playlist_genre": le.classes_,
    "label_encoded_value": le.transform(le.classes_)
})
print(mapping_df)
```

```python
# One Hot Encoding (Encoding playlist_genre)

df_onehot = pd.get_dummies(df["playlist_genre"])
df_onehot
```

```python
# Ordinal Encoding (Encoding Grade)

data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Grade': ['A', 'B', 'C', 'A', 'B']
}
df = pd.DataFrame(data)
encoder = OrdinalEncoder(categories=[['A', 'B', 'C']])
df['Grade_encoded'] = encoder.fit_transform(df[['Grade']])
print(df)
```
