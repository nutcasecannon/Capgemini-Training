```python
import pandas as pd
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    "feature_1": [12, 24, 5],
    "feature_2": [5, 20, 18],
    "feature_3": [2, 5, 10]
}
df = pd.DataFrame(data)
```

```python
# Normalization

Norm = Normalizer()
data = Norm.fit_transform(df)
df_norm = pd.DataFrame(data,columns=df.columns)
df_norm
```

```python
# Standard Scaler

scaler = StandardScaler()
data = scaler.fit_transform(df)
df_scaler = pd.DataFrame(data, columns=df.columns)
df_scaler
```
